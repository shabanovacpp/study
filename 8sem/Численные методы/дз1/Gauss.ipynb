```python
import numpy as np
from read_file import parse_variant
from print_matrix import print_matrix
from print_matrix import print_numbers
np.set_printoptions(suppress=True, precision=16)

var = 19
```

```python
def Gauss(A, b):

    n = A.shape[0]       # Определяем порядок СЛАУ
    x = np.zeros(n)      # 

    #Прямой ход
    for k in range(n):
        # Нахождение главного элемента по столбцу
        a_mk = np.argmax(abs(A[k:,k])) + k # В каждой строке берем нулевой элемент и среди них находим индекс максимального (определяем номер строки, в которой максимальный коэффициент перед xk)
        if k != a_mk:
            #Перестановка строк 
            A[[k, a_mk]] = A[[a_mk, k]]
            b[[k, a_mk]] = b[[a_mk, k]]
        if (A[k][k] == 0):
            return f"a[{k}][{k}] = 0! Invalid matrix"
        temp = A[k][k]
        A[k] = A[k] / temp         # нормируем строку по a[k][k]
        b[k] = b[k] / temp

        for j in range(k+1, n):
            temp = A[j][k]
            A[j] = A[j] - A[k] * temp
            b[j] = b[j] - b[k] * temp

    # Обратный ход
    for i in range(n-1, -1, -1):
        x[i] = b[i]
        for j in range(i+1, n):
            x[i] = x[i] - A[i][j] * x[j]

    return x
```

<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Вычисление невязки
</div>

$$r = b - Ax^*$$

```python
def Residual(A, b, x):

    return b - np.dot(A,x)
```

<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Вычисление единичной нормы
</div>
<br>
<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Для вектора: 
</div>

$$\|x\|_1 = \sum_{i=1}^{m} |x_{i}|$$

 <br>
<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Для матрицы: 
</div>

$$\|A\|_1 = \max_{1 \leq j \leq n} \sum_{i=1}^{m} |a_{ij}|$$



```python
def Norm1(a):
    if a.ndim == 1:
        return np.sum(np.abs(a))
    else:
        return np.max(np.sum(np.abs(a), axis=0))
```

<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Вычисление бесконечной нормы
</div>
<br>
<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Для вектора: 
</div>

$$\|x\|_\infty = \max_{1 \leq i \leq m} |x_{i}|$$

 <br>
<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Для матрицы: 
</div>

$$\|A\|_\infty = \max_{1 \leq i \leq n} \sum_{j=1}^{m} |a_{ij}|$$

```python
def NormInf(a):
    if a.ndim == 1: 
        return np.max(np.abs(a))
    else:
        return np.max(np.sum(np.abs(a), axis=1))  
```

```python
def Error(true, real):
    abs_error_1 = Norm1(true - real)                       # абсолютная погрешность
    rel_error_1 = abs_error_1/Norm1(true)                  # относительная погрешность

    abs_error_inf = NormInf(true - real)                   # абсолютная погрешность
    rel_error_inf = abs_error_inf/NormInf(true)            # относительная погрешность

    return abs_error_1, abs_error_inf, rel_error_1, rel_error_inf
```

<div style="font-family: 'Times New Roman', serif; font-size: 24px; text-align: center;">
    Вычисление числа обусловленности
</div>

$$v(A)=\|A^{-1}\|\cdot\|A\|$$

```python
def Cond(matrix):
    cond_1 = np.dot(Norm1(matrix), Norm1(np.linalg.inv(matrix)))
    cond_inf = np.dot(NormInf(matrix), NormInf(np.linalg.inv(matrix)))
   
    return cond_1, cond_inf
```

```python
file_path = 'Данные_ЛР_01,_03_Гаусс,_итерационые_методы.txt'
A_good, b_good, x_good_true, A_bad, b_bad, x_bad_true = parse_variant(file_path, var)
print_matrix(A_good, b_good.reshape(-1,1), A_bad, b_good.reshape(-1,1), titles=['A (good)', 'b (good)', 'A (bad)', 'b (bad)'])  # вывод расширенных матриц
print("\n\n")
print_matrix(x_good_true, x_bad_true, titles=['x (good)', 'x (bad)'], spaces=30)   # вывод истинного решения

x_bad = Gauss(A_bad, b_bad)     # решаем хорошо обусловленное СЛАУ
x_good = Gauss(A_good, b_good)  # решаем плохо обусловленное СЛАУ

print()
print_matrix(x_good, x_bad, titles=['x* (good)', 'x* (bad)'], spaces=30)     # вывод нашего вселенного
```

```python
abs_error_1_good, abs_error_inf_good, rel_error_1_good, rel_error_inf_good = Error(x_good_true, x_good)
abs_error_1_bad, abs_error_inf_bad, rel_error_1_bad, rel_error_inf_bad = Error(x_bad_true, x_bad)
print_numbers(abs_error_1_good, abs_error_1_bad,
              titles=['Абсолютная погрешность (good, 1)', 'Абсолютная погрешность решения (bad, 1)'])
print_numbers(abs_error_inf_good, abs_error_inf_bad,
              titles=['Абсолютная погрешность (good, ∞)', 'Абсолютная погрешность решения (bad, ∞)'])
print()
print_numbers(100*rel_error_1_good, 100*rel_error_1_bad,
              titles=['Относительная погрешность (good, 1), %', 'Относительная погрешность решения (bad, 1), %'])
print_numbers(100*rel_error_inf_good, 100*rel_error_inf_bad,
              titles=['Относительная погрешность (good, ∞), %', 'Относительная погрешность решения (bad, ∞), %'])
```

```python

r_good = Residual(A_good, b_good, x_good)   
r_bad = Residual(A_bad, b_bad, x_bad)       # выычисление невязки

print_numbers(Norm1(r_good), Norm1(r_bad), titles=['Единичная норма невязки (good)', 'Единичная норма невязки (bad)'])
print()
print_numbers(NormInf(r_good), NormInf(r_bad), titles=['Бесконечная норма невязки (good)', 'Бесконечная норма невязки (bad)'])
```
